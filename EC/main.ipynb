### Теоретическое введение

#### Электрокапиллярные кривые

Если на поверхность металла, находяшегося под слоем водного раствора, нанести каплю неполярной органической жидкости (в нашем случае - декана), то на трехфазной границе установится равновесие согласно уравнению Юнга (1), где m - металл, l - раствор, o - органическая жидкость:

$$ \sigma_{ml} = \sigma_{mo} + \sigma_{ol} \cdot cos \;\theta \qquad (1)$$

При изменении разности потенциалов между ртутью и водным раствором межфазное натяжение $\sigma_{ml}$ меняется в соответствии с уравнением Липпмана (2):

$$ q = - \frac{\partial \sigma}{\partial E} \qquad (2)$$

Которое следует из фундаментального уравнения Гиббса (3):

$$ d \sigma = -q dE - \sum{\Gamma_i d\mu_i} \qquad (3)$$
$$ d \sigma = -q dE - RT \sum{\Gamma_i d (ln \; c_i) }$$

Также из него получить второе уравнение Липпмана:

$$ C = \frac{\partial q}{\partial E_{c_i}} = \frac{\partial^2 \sigma}{\partial E^2_{c_i}} \qquad (4)$$

Интегрируя это уравнение в рамках простейшей модели Гельмгольца, где емкость двойного электрического слоя не зависит от разности потенциала (выполняется при высоких концентрациях электролита/вдали от ПНЗ), получаем следующее выражение:

$$ \sigma - \sigma_0 = -C \frac{(E - E_0)^2}{2} \qquad (5)$$

из которого очевидно, что электрокапиллярная кривая должна иметь вид параболы.

### Представления о строении двойного электрического слоя
**Модель Гельмгольца:**

Самой простой моделью является модель Гельмгольца, в которой противоионы находятся в монослое, нейтрализуя заряд электрода. Тогда это можно представить как плоский коденсатор, емкость которого вычисляется по формуле (6):

$$ C = \frac{\epsilon \epsilon_0}{d} \cdot S \qquad (6)$$

где $\epsilon$ - диэлектрическая проницаемость, $\epsilon_0$ − ($8.85 \cdot 10−12 \; Ф/м$, $d$ - расстояние между обкладками, фактически равное радиусу притянутого иона)

**Модель Гуи-Чапмена:**

Для этой модели делается несколько предположений:
1) Ионы = точечные заряды
2) Однородная среда
3) Ионы относительно электрода распределены по уравнению Больцмана

По теории Гуи-Чапмена,

$$ C = \sqrt{\frac{R \epsilon \epsilon_0 c F^2}{RT}} sh(\frac{z F \phi}{2RT}) \qquad (7)$$

При этом важно учитывать, что в плотном слое диэлектрическая проницаемость воды сильно уменьшена, т.к. она жестко ориентирована в пространстве.


$$ \lambda \sim \frac{1}{c} $$

Недостатком теории Гуи-Чапмена является то, что емкость двойного слоя получалась на несколько порядков больше опытных значений. Этот результат был связан с предположением о том, что ионы являются математическими точками, а значит, могут бесконечно приближаться к поверхности электрода.

**Модель Штерна:** 

Модель Штерна является сочетанием моделей Гуи-Чапмена и модели Гельмгольца + в ней учитывается специфическая адсорбция. В ней вводится внутренняя плоскость Гельмгольца, где локализуются центры специфически адсорбированных ионов, а также внешняя плоскость - граница максимального приближения к поверхности центра сольватированных ионов. Согласно объединенной модели,

$$ \frac{1}{C_{total}} = \frac{1}{C_Г} + \frac{1}{C_D} \qquad (8)$$

#### Исследование электрокапиллярной кривой на ртути с помощью измерения краевого угла смачивания декана

Нанесем каплю декана объемом примерно 5 мкл на поверхность ртути, погрузив кончик дозатора в ртуть. Варьируя потенциал ртутного электрода в диапазоне (300; -1300) мВ в циклическом режиме со скоростью развертки около -50 мВ/c (3-5 циклов), проведем тренировку капли.

Произведем измерения краевого угла смачивания от величины потенциала ртутного электрода в диапазоне от -1300 до 300 мВ, снимая показания каждые 100 мВ. С помощью программы *ImageJ* определим краевой угол у полученных снимков и найдем $\sigma_{ml}$ по формуле (1), используя табличные данные $\sigma_{Hg|decane}$ = 375 мН/м, $\sigma_{H_2O|decane}$ = 51 мН/м. График $\sigma(E)$ хорошо аппроксимируется полиномом второй степени (из ур-ия (5)).

```python
sigma_hg_decane = 375
sigma_decane_h2o = 51
```

```python
angle = 180 - np.array([88.72, 97.66, 122.83, 135.51, 145.77, 150.75, 146.85, 157.27, 
         158.20, 156.40, 155.41, 161.01, 158.24, 139.55, 143.16, 119.90, 87.18])
angle = np.deg2rad(angle)

E = np.arange(-1300, 301, 100)
```

```python
fig, axs = plt.subplots(figsize=(15,10))

x = E
y = sigma_hg_decane + sigma_decane_h2o * np.cos(angle)
axs.scatter(x, y, marker='v', c='r')

a = np.polyfit(x, y, 2)
f = np.poly1d(a)
axs.plot(np.linspace(-1300, 300, 1000), f(np.linspace(-1300, 300, 1000)), label=f"Аппроксимирующий полином чевтвертой степени")

axs.set_xlabel("E, мВ", fontsize=14)
axs.set_ylabel(r"$\sigma, мДж/м^2$", fontsize=14)
axs.set_title("Электрокапиллярная кривая ртути в водном растворе 0.1М NaF", fontsize=18)

axs.grid()
plt.legend(loc='best', fontsize=15)
plt.tight_layout()
plt.show()
```

Отсюда легко вычислить ПНЗ и коэффициент поверхностного натяжения чистой ртути: 
$$ПНЗ = argmax(f(I)) = -458 \; мВ $$    
$$\; \sigma_{Hg} = max(f(I)) = 427 \; мН/м $$ 

Отсюда же можно получить оценку величины емкости, которая равна:
$$С = -\frac{\partial^2 \sigma}{\partial E^2} = -2a \approx 145.2 \; мФ/м^2$$

Упрощённо считая ДЭС плоским конденсатором, можно найти расстояние между его обкладками по формуле ($\epsilon_{H_2O} = 78.3$):

$$ d = \frac{\epsilon \epsilon_0}{C} \approx 4.78 \; нм$$ 

```python
def csvHandler(filename, title=None):
    plt.figure(figsize=(15,10))
    
    df = pd.read_csv(f"{filename}", sep=r"\s+", header=None)
    df.drop(df.columns[[0, 1]], axis=1, inplace=True)
    df.columns = ['I', 'U']
    df = df.dropna()

    x = 1000*df['I']
    y = 1000*df['U']
    
    plt.xlabel("I, mA", fontsize=15)
    plt.ylabel("U, mV", fontsize=15)
    plt.title(title, fontsize=18)
    
    plt.grid()
    plt.tight_layout()
    plt.plot(x, y)
```
