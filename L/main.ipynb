### Теоретическое введение 

**Дисперсная система** - образование из двух или более фаз, которые практически не смешиваются и не реагируют друг с другом химически. Расмотрим дисперсную систему из двух фаз. То вещество, которое присутствует в меньшем количестве и распределено в объеме другого, называется дисперсной фазой. Вещество, присутствующее в большем количестве, в объеме которого распределена дисперсная фаза, называется дисперсионной средой. По величине частиц веществ, составляющих дисперсную фазу, дисперсные системы делятся на: 

1. **Грубодисперсные** (взвеси) с размерами частиц более 100 нм;
2. **Тонкодисперсные** (коллоидные растворы или коллоидные систе- мы) с размерами частиц от 1 до 100 нм;
3. **Гомогенные** системы, или растворы с размером частиц менее 1 нм. Такая система однородна, поверхности раздела между частицами и средой нет.

Коллоидные системы подразделяются на две группы, отличающиеся по характеру взаимодействий между частицами дисперсной фазы и дисперсионной среды (классификация Ребиндера): 
  
1. **Лиофильные** коллоидные растворы - частицы дисперсной фазы хо- рошо взаимодействуют с дисперсионной средой (большое сродство к растворителю), низкое значение поверхностного натяжения, термодинамически устойчивы и образуются самопроизвольно;
   
2. **Лиофобные** коллоидные растворы (золи) - частицы дисперсной фазы слабо взаимодействуют с дисперсионной средой. Термодинамически неустойчивые к агрегации системы, образуются несамопроизвольно в результате диспергирования или конденсации с пересыщением. Система обладает избытком поверхностной энергии (=> самопроизвольно протекают процессы укрупнения частиц); эти системы могут быть получены только с затратой энергии и устойчивы лишь в присутствии стабилизаторов. Однако они могут быть устойчивы кинетически. Для того, чтобы в среде появились частицы фазы, надо создать зародыши, на которые будут осаждаться молекулы фазы (для уменьшения поверхностного натяжения). На создание зародышей тратится энергия, поэтому без затраты энергии частицы не образуются. Но самопроизвольно полученные частицы фазы в среде начнут слипаться (коагулировать) и в итоге выпадут в осадок, поэтому система устойчива только в присутствии каких-либо стабилизаторов.

В данной работе рассматриваются лиофобные коллоидные системы - частицы труднорастворимых осадков неогранических соединений. Для примера рассмотрим получение $AgCl$ в избытке $KCl$: $$AgNO3 + KCl_{изб} = AgCl + KNO3$$ 

На поверхности твердого $AgCl$ адсорбируются ионы $Cl^−$, в результате получается отрицательно заряженый золь хлорида серебра. Заряд золя определяется зарядом адсорбированных ионов.

**Правила избирательной адсорбции:**

1. Если в растворе присутствуют ионы в одинаковой концентрации, то адсорбироваться будет тот, у кого заряд больше;

2. Если в растворе присутствуют ионы с одинаковым зарядом, то адсорбироватся будет тот, у которого концентрация больше;

3. Если в растворе присутствуют ионы с одинаковыми зарядами, в одинаковой концентрации, то по правилу Фаянса-Пескова-Панета на поверхности твёрдого вещества будут адсорбироваться ионы, которые могут достраивать кристаллическую решётку (входят в её состав или изоморфны) или те ионы, которые образуют труднорастворимые соединения с ионами кристаллической решётки.

4. Хорошо адсорбируются ионы, которые сильнее деформируются под действием поля кристаллической решётки(тем самым уменьшая полярность связи). Хорошо адсорбируются большие органические ионы, так как они обладают большей поляризуемостью и меньше гидратируются.

5.  Из ионов с одинаковыми зарядами адсорбируется лучше тот, который меньше гидратирован(адсорбционная способность возрастает в лиотропных рядах Гоффмейстера). Если ионы противоположных зарядов, то лучше адсорбируются катионы.

**Строение двойного электрического слоя и $\phi$-потенциал.** 

Ионы, которые избирательно адсорбируются на коллоидной частице, создают $\phi$-потенциал. На практике его почти невозможно измерить, однако есть другой способ определения заряда частицы: определение скорости частицы в электрическом поле. При таком движении ДЭС разрывается и место разрыва называется плоскостью скольжения (зачастую граница между плотной и диффузной частью ДЭС). Потенциал плоскости скольжения $\zeta$- потенциал, который возможно измерить, он равен по абсолютной величине разности потенциалов дисперсионной среды и неподвижной жидкости, окружавшей частицу. Важно, что значение $dzeta$-потенциала может быть связано с устойчивостью коллоидов(при $|\zeta| > 30 \; мВ$, коллоиды считаются устойчивыми)

**Теория ДЛФО**   

Частицы фазы заряжены (плотный слой на поверхности труднорастворимой соли), поэтому они взаимодействуют друг с другом. Согласно теории ДЛФО, взаимодействие между частицами среды ($r$ - расстояние между частицами): 
- притяжение - силы Ван-Дер-Ваальса, $U \sim 1/r^2$
- отталкивание - электростатические взаимодействия $U \sim 1/e^r$

![рис.2](image2.png)

**Рис. 2: Энергия взаимодействия. 1 - энергия отталкивания, 2 - энергия притяжения, 3 - суммарная энегрия взаимодействия**

Первый минимум на графике - две частицы при непосредственном контакте. На средних расстояниях между частицами энергия отталкивания преобладает над энергией притяжения, поэтому возникает потенциальный барьер, препятствующий агрегации. Для того, чтобы устранить барьер, в раствор вводятся электролиты. Введенный электролит может действовать на частицы 2 способами: 

1. Ионы электролита адсорбируются на коллоидной частице, понижая при этом ее заряд. В этом случае отталкивание между частицами уменьшается, следовательно понижается энергетическиий барьер и частицы начинают личше коагулировать. Такая коагуляция называется **нейтрализационной**.
 
2. Коллоидная частица окружена диффузным ДЭС, который зависит от ионной силы раствора $I = \frac{\sum{c_i z^2_i}}{2}$ следующим образом (согласно теории Гуи-Чапмена):

   $$ \lambda = \sqrt{\frac{\epsilon \epsilon_0 R T}{2 F I^2}} \qquad (1)$$
 
   Значит, при добавлении электролита увеличивается ионная сила раствора и уменьшается толщина диффузного слоя ДЭС. Это позволяет      частицам ближе подойти друг к другу и коагулировать. Такой тип коагуляции называется **концентрационным**.

При больших зарядах коллоидных частиц проявляется концентрационная коагуляция, а при малых - нейтрализационная. Таким образом, при увели- чении концентрации электролита энергетический барьер уменьшается. Концентрация электролита, при которой энергетический барьер исчезает, называется **порогом быстрой коагуляции**

$$ C_{ПК} = \frac{C_e}{V_з + V_e} \qquad (2)$$

где $C_е$, $V_е$ - концентрация и объем добавленного электролита, $V_з$ - объем золя в растворе. Порог быстрой коагуляции может зависеть от заряда иона электролита - противоиона к коллоиднолй частице $z$, следующим образом:

- **правило Дерягина-Ландау**: при электролитной коагуляции по концентрационному механизму:
  $$ C \sim \frac{1}{z^6} $$
- **правило Эйлерса-Корфа**: при электролитной коагуляции по нейтрализационному механизму:
  $$ C \sim \frac{1}{z^2} $$

**Кинетика коагуляции**  

Рассмотрим лиофобоную коллоидную систему. Термодинамически системе выгодно коагулировать, но из-за существования большого потенциального барьера коагулировать может быть не выгодно кинетически. Быстрая коагуляция - потенциальный барьер отсутствует (например, он иcчез при добавлении электролита). **Теория скорости быстрой коагуляции Смолуховского:** 

Предположения: 
1. Потенциальный барьер отсутствует, поэтому любое столкновение частиц приводит к слипанию;
2. Скорость соударений определяется диффузией;
3. В 1 акте соударения участвуют ровно 2 частицы;
4. Частицы сферические.
  


Пусть $n$ - концентрация коагулирующих частиц в момент времени $t$ (в $м^{−3}$), $k$ - константа скорости коагуляции
$$ \frac{dn}{dt} = -kn^2 \qquad (3)$$
$$ \int_{n_0}^{n} \frac{dn}{n} = -k \int_{0}^{t} dt $$
$$ \frac{1}{n} - \frac{1}{n_0} = kt $$
$$ n(t) = \frac{n_0}{1 + k n_0 t} $$

Так как скорость соударений определяется диффузией, то из закона Фика скорость реакции ($d$ - диаметр частицы, $D$ - коэффициент диффузии):
$$ -kn^2 = V = jSn = 4 \pi d^2 D \frac{n}{d} n$$
$$ k = 8 \pi r D$$

Из формулы Эйнштейна ($\eta$ - вязкость жидкости):
$$ D = \frac{RT}{6 \pi N_A \eta r} \qquad (4)$$

Тогда константа быстрой коагуляции:
$$ k = \frac{4RT}{3 N_A \eta} = \frac{4KT}{3 \eta} \qquad (5)$$

В случае медленной коагуляции (есть потенциальный барьер):
$$ k = \frac{4KT}{3 \eta} exp(\frac{\Delta E}{RT}) \qquad (6)$$

#### Получение золя серы методом замены растворителя

Нальем в пробирку около 5 мл дистилированной воды и добавим туда несколько капель насыщенного раствора серы в ацетоне. Примерно через две минуты раствор начнет опалесцировать. Опалесценция - явление, при котором свет рассеивается, проходя через неоднородную среду. В данном случае, скорее всего, имеет место рэлеевское рассеяние (на частицах много меньше длины волны), при котором рассеивается синий свет, а проходит - желтый).

Таким образом, в данном случае мы заменяем ацетон (органический неполярный), в котором сера растворима, на дистилированную воду (полярная), в которой сера практически не растворяется. Происходит пересыщение, и образуется золь.

#### Получение золя гидроксида железа $Fe(OH)_3$ методом конденсации

В стакан из термостойкого стекла нальем 50 мл дистилированной воды и, закрыв его часовым стеклом во избежание слишком сильного испарения, доведем воду до кипения на электроплитке с мешалкой. Не снимая воду с электроплитки, медленно, при перемешивании зальем 5 мл 2% -ного раствора $Fe(OH)_3$. В результате образуется раствор коньячного цвета.

Происходящая реакция:
$$ FeCl_3 + 3H_2O = Fe(OH)_3 + 3HCl$$

При этом в растворе по следующей реакции образуется оксохлорид железа:
$$ Fe(OH)_3 +HCl = FeOCl + 2H_2O $$
$$ FeOCl = FeO^+ + Cl^−$$

Таким образом, в растворе присутствуют ионы $Fe^{3+}$, $Cl^−$, $FeO^+$. По правилу Фаянса-Пескова-Фанета на поверхности коллоидных частиц будут лучше адсорбироваться те молекулы, которые достраивают кристаллическую решетку вещества. К тому же, лучше адсорбируются ионы с большим зарядом. Под оба критерия подходят ионы $Fe^{3+}$, $FeO^+$, поэтому именно они, а не ион хлорида, будут служить стабилизатором. Таким образом, формула мицеллы золя будет следующей:

$$ \{{m[Fe(OH)_3]nFeO^+(n − x)Cl^−}\}^{x+}Cl^− $$

Или, альтернативно:

$$ \{{m[Fe(OH)_3]nFe^{3+}(3n − x)Cl^−}\}^{3x+}Cl^− $$

В этой мицелле агрегат состоит из $m$ частиц $Fe(OH)_3$, ядро - из агрегата и $nFeO^+$, а частица и мицелла - из ядра и $Cl^−$

#### Отрицательно заряженный золь золота

В стакан отмерим 25 мл дистиллированной воды и на плитке доведем до кипения. К полученному раствору добавим 250 мкл раствора $H[AuCl4]$ с концентрацией 1% (по массе). Через 3 минуты кипячения добавим 300 мкл раствора $Na_3Cit$ концентрацией 1% (по массе). Затем смесь кипятим примерно 5 минут (до того момента, как цвет раствора перестанет меняться). Полученную смесь снимем с плитки и оставим остывать до комнатной температуры.

Суммарную реакцию цитратного восстановления можно записать как: 

$$ 2H[AuCl_4]+ 3Na_3HOC(COO)(CH_2COO)_2 \; \rightarrow \; 2Au + 3Na_2O = C(CH_2COO)_2 + 3CO_2 + 3NaCl + 5HCl $$

Из стехиометрии данной реакции цитрат-ионы взяты в избытке, поэтому можно предположить, что именно они адсорбируются на частице золота. С другой стороны, в растворе много ионов $Cl^−$, которые могут также адсорбироваться на частице. Таким образом, предполагаемый вид получившейся мицеллы:

$$ \{[mAu] n(O = C − (CH_2COO)_2))^{2−} \; kCl− (2n+ k − x)Na^+\}^{x−} xNa+$$

#### Определение порогов коагуляции золей

В данном опыте будем определять пороги коагуляции золя золота из 1 части работы при электролитной коагуляции с помощью электролитов, дающих различные заряды противоионов к коллоидной частице. Так как полученная мицелла золота выглядит следующим образом:

$$ \{[mAu]n Cyt^{3-} (3n-x) Na^+\}^{x-} xNa^+$$

то противоионы - катионы. Рассмотрим 3 соли: KNO_3,Mg(NO_3)_2, La(NO_3)_3 с зарядами катионов +1, +2, +3 соответственно. Определим для каждой из них порог быстрой коагуляции следующим методом: наливаем в пробирку 2 мл раствор электролита разной концентрации и 1 мл раствора золя золота. По времени изменения цвета опредляем тип коагуляции (быстрая / медленная). В зависимости от результата определяем следующую нужную концентрацию электролита (аналогично бинарному поиску).

![рис. 5](image4.png)

Таблица 2: Коагуляция в зависимости от концентрации добавленного электролита. $+$ означает быструю коагуляцию, $-$ ее отсуствие, $\pm$ медленную

Из этого находим пороговые концентрации:

$$ C_1 = 0.0583 \; M $$
$$ C_2 = 0.0183 \; M $$
$$ C_3 = 0.00483 \; M $$

Логарифмируя выражение для порога коагуляции $ C_{ПК} = \frac{1}{z^k} $ получаем $ln(C_{ПК}) = ln(A)− kln(z)$  
Построим график зависимости $ln(C_{pk})$ от $ln(z)$:

```python
C = np.array([0.0583, 0.0183, 0.00483])
z = np.array([1, 2, 3])
```

```python
plt.figure(figsize=(15,10))

x = np.log(z)
y = np.log(C)
plt.scatter(x, y, marker="v", c="r")

a = np.polyfit(x, y, 1)
f = np.poly1d(a)
plt.plot(x, f(x))

plt.xlabel(r"$ln(z)$", fontsize=15)
plt.ylabel(r"$ln(C_{ПК})$", fontsize=15)

plt.grid()
plt.title(r"Зависимость $ln(C_{ПК})$ от $ln(z)$", fontsize=18)
plt.show()
```

Таким образом, $k \approx 2.2$, $C_{ПК} \sim \frac{1}{z^{2.2}}$ . Значит, происходила как нейтрализационная, так и концентрационная коагуляции. Так как 2.2 ближе к 2, чем к 6, то можно предположить, что в большей мере проявлялся нейтрализационный механизм.

Пороги коагуляции приблизительно относятся друг к другу как 1 : 4 : 12, что достаточно неплохо соответствует правилу Эйлерса-Корфа (1 : 4 : 9)
