### Теоретическое введение
#### Химический потенциал и активности компонентов раствора
Химический потенциал компонента реального раствора вводится следующим образом:
$$ \mu = \mu_0 + RTln \; a, \qquad(1) $$
где $\mu_0$ - химический потенциал этого компонента раствора в его "стандартном" состоянии. Таким образом, $a$ вводится как безразмерный коэффициент, отражающая вклад компонента в свободную энергию Гиббса раствора. Активность ai - безразмерная величина, которая представляется как
$$ a_i = \gamma_i · \chi_i, \qquad (2) $$
где $\chi_i$ - мольная доля, а $\gamma$ - коэффициент активности
Для растворителя $А$ в качестве стандартного состояния принимается чистый растворитель, причем $\gamma \rightarrow 1$
при мольной доле, стремящейся к единице А для растворенных веществ, наоборот, коэффициент активности
стремится к единице при бесконечном разбавлении раствора

#### Теория Дебая-Хюккеля
Теория Дебая-Хюккеля позволяет представить коэффициент активности $\gamma_i$ иона в виде разницы энергии иона
типа $i$ в реальном растворе ионной силы $I$ и в предельном случае, когда ионы вокруг него отсутствуют (I = 0)
$$ RTln\gamma_i = N_a · (U_I − U_{I=0}) \qquad (3) $$
Распределение потенциала вокруг некоторого иона в приближении Дебая-Хюккеля получается следующим:
$$ \phi(r) = \frac{z_ie}{4π\epsilon\epsilon_0r}e^{−\chi·r}, \qquad (4) $$
что качественно интерпретируется как образование экранирующей ионной атмосферы, где $\chi$ - величина, обратная радиусу Дебая-Хюккеля Отсюда вытекает предельный закон Дебая-Хюккеля
$$ lg \; \chi_i = −{z_i}^2 h \sqrt I \qquad (5)$$

Для водных растворов при $25°C$ $h$ можно принять за 0.51. Измерить активность ионов одного типа не удастся, поскольку необходимо выполнять условие электронейтральности. Поэтому для системы принято говорить о среднеионном коэффициенте активности. В первом приближении теории Дюбая-Хюккеля среднеионный коэффициент при $25°C$ выражается по следующей формуле (верна только при $I <$ 0.01 M):
$$ lg \; \gamma_{+−} = −0.51|z_+ z_-| \sqrt I \qquad (6)$$
Во втором учитывается, что центры ионов не могут прилижаться друг к другу на растояние меньше некоторого $a$:
$$ lg \; \gamma_{+−} = −0.51|z_+ z_−| \frac{\sqrt I}{1 + \sqrt I} \qquad (7) $$
Эта формула хорошо работает при $25°C$ и $I <$ 0.1 M. В третьем приближении приходится учитывать добавочный коэффициент, поскольку изменение коэффициента активности может стать нелинейным. (Рис. 1)
$$ lg \; \gamma_{+−} = −0.51|z_+ z_−| \frac{\sqrt I}{1 + \sqrt I} + bI \qquad (8) $$

#### О потенциометрическом определении pH
Хороший pH-метр = милливольтметр, с помощью которого проводятся измерения разности потенциалов между электродами, погруженными в раствор. Разность потенциалов линейно связана исключительно с pH, без активностей каких-либо других ионов.
$$ E_X = E_0 − b · pH \qquad (9) $$
Неизвестные $b$, $E_0$ можно найти по калибровке по буферным растворам (эксперимент 3.1) Если концентрация протонов известна, то можно найти коэффициент активности в растворе:
$$ γ_{H^+} = \frac{10^{E_0 − E_X / b}} {C_{H^+}} \qquad (10) $$

#### Диффузионный потенциал
Диффузионный потенциал возникает из-за разницы в подвижности ионов. Происходит разделение положительных и отрицательных ионов, что приводит к возникновению скачка электрического потенциала, называемого в данном случае дифузионным потенциалом $\Delta \phi_D \;$. 
Точной формулы, позволяющей расчитать дифузионный потенциал в общем случае, нет, поэтому его значение всегда вносит неопределенную погрешность в измерения. Для уменьшения его значения используют солевые мостики с растворами, чьи ионы имеют примерно одинаковую подвижность, например, $KCl$   

#### Стеклянный электрод
Если активность $H^+$ преобладает над вкладом ионов $Na^+$ (выполняется в кислых/нейтральных средах), то для потенциала стеклянного электрода можно использовать следующую формулу:
$$ \phi = \phi_0 + \frac{RT}{F} ln \; a_{H^+} \qquad (11) $$
При большой концентрации $Na^+$ в щелочной среде могут возникать значительные отклонения функции $E(pH)$ от линейной зависимости. Это называется щелочной (катионной) ошибкой стеклянного электрода.

#### Гальваническая ячейка
Для независимости потенциала электрода от состава исследуемого раствора составляют гальваническую ячейку, например, такую, как на рис. 2. Для нее можно использовать формулу:
$$ E_x = E_0 − \frac{RT ln \; 10}{F} pH_x, \qquad (12) $$ 
где $E_x$ - ЭДС цепи в растворе $X$.    
Также pH в средах в заданной концентрацией хлорид-ионов можно исследовать через среднюю активность $\sqrt{a_{H^+} a_{Cl^−}}$ 
$$ E_x = E_0 + \frac{RT}{F} ln \; a_{H^+} a_{Cl^−}, \qquad (13) $$
Если известно, например, что концентрации $c_{H^+} = c_{Cl^−}$, можно использовать полученные результаты для определения изменения активностей ионов гидроксония.

### Выводы
С помощью калибровки по буферным растворам для pH-метра были найдены коэффициенты, которые были применены при дальнейших расчетах активности и коэффициентов активности электролитов. С помощью электродов с двумя разными типами соединений определен средний коэффициент активности $HCl$ и проведено сравнение результатов.   

В ячейках с жидкостным соединением в месте контакта двух отличающихся по составу растворов возникает диффузионный потенциал, точное значение которого расчитать не получится, поэтому для определения среднего коэффициента активности электролитов правильнее будет использовать электроды без жидкостного соединения.
Через удельную электропроводность раствора была найдена константа диссоциации кислоты, которая достаточно близко сходится с теоретической.
