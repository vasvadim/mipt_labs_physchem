### Теоретическое введение  
#### Типы поверхностно активных веществ и их основные свойства
На основании уравнения Гиббса для бинарных растворов при сравнительно низкой концентрации:
$$ \Gamma = -\frac{c}{RT} \cdot \frac{d\sigma}{d c} \; , \qquad (1)$$
поверхностную активность растворенных веществ количественно определяют по значению производной
$ g = −d \sigma /dc$ при $ c \rightarrow 0 $. **Поверхностно-активные вещества** в соответствии с уравнением Гиббса понижают межфазное натяжение и для них $ g > 0 $. **Поверхностно-инактивные вещества**, напротив, повышают натяжение и для них $ g < 0 $. Экспериментально поверхностную активность определяют по тангенсу
угла наклона касательной к графику зависимости натяжения от концентрации (рис. 1) растворенного
вещества в точке ее пересечения с осью ординат.

![рис.1](image1.png)
##### <center> Рис. 1: Графики зависимости поверхностного натяжения от концентрации <center/>

По способности к заряжению полярной части в водных растворах все ПАВ делят на **ионогенные**
и **неионогенные**. Ионогенные ПАВ подразделяют на **анионные** (заряженные отрицательно), **катионные** (заряженные положительно) и **амфолитные**, полярная часть которых несет два противоположных
заряда.

#### Понятие о критической концентрации мицеллообразования   
Термодинамически устойчивые (лиофильные) самоорганизованные гетерогенные дисперсные системы,
образуемые молекулами ПАВ являются **мицеллярными лиофильными коллоидными растворами**.
Благодаря высокой поверхностной активности молекулы ПАВ можно использовать также для **солюбилизации** лиофобных коллоидных систем. Для самопроизвольного образования мицелл в растворе ПАВ
концентрация раствора должна превысить некоторое пороговое значение, называемое **критической концентрацией мицеллообразования** (ККМ). Ниже этой концентрации в растворе находятся главным образом одиночные молекулы и раствор может считаться истинным, а выше, после формирования мицелл,
– коллоидным. Число молекул, составляющих мицеллу, называют числом агрегации m, а общую ее массу
– мицеллярной массой. Мицеллы – не статичные образования, а находятся в постоянном динамическом
равновесии с составляющими их одиночными молекулами ПАВ.

### Практические методы определения ККМ
Методы определения ККМ основаны на резком изменении различных физико-химических свойств
растворов ПАВ (поверхностного натяжения, мутности, удельной или эквивалентной электропроводности,
показателя преломления, осмотического давления, вязкости и т.д.) при переходе от молекулярного раствора к мицеллярному. При таком переходе с ростом общей концентрации ПАВ концентрация отдельных
молекул перестает быстро увеличиваться, т.к. дополнительно поступающее вещество идет на формирование многомолекулярных мицелл.

#### Кондуктометрический метод
Кондуктометрическое определение ККМ основано на измерении концентрационной зависимости электропроводности растворов ионогенных ПАВ. Если бы в водных растворах ионогенных ПАВ отсутствовало
мицеллообразование, то в согласии с уравнением Кольрауша, экспериментально измеренная эквивалентная электропроводность в зависимости от квадратного корня из концентрации описывалась бы линейной функцией. Это выполняется лишь при малых концентрациях ПАВ, а начиная с ККМ формируются
ионные мицеллы, окруженные диффузным слоем противоионов, и на экспериментальной зависимости наблюдается излом. Подвижности ионных мицелл меньше подвижности ионов. Кроме того, значительная
часть противоионов находится в плотном адсорбционном слое, что существенно уменьшает электропроводность раствора ПАВ (заряд мицелл меньше суммарного заряда составляющих их молекул ионогенного
ПАВ). Поэтому при увеличении концентрации ПАВ выше ККМ наблюдается уменьшение скорости роста
удельной электропроводности (рис. 3).

Определение ККМ по изменению поверхностного натяжения
раствора ПАВ
Поверхностное натяжение водных растворов ПАВ уменьшается с ростом концентрации вплоть до ККМ
(рис. 4). Изотерма поверхностного натяжения, $\sigma = f(ln \; c_{ПАВ})$ в области низких концентраций ПАВ имеет 
криволинейный участок, на котором в соответствии с уравнением адсорбции Гиббса адсорбция Γ на
межфазной границе возрастает с ростом концентрации:
$$ -d\sigma = RT \cdot \Gamma d \; ln \; c_{ПАВ} \qquad (2)$$
При определенной пороговой концентрации $c_m$ криволинейный участок изотермы переходит в линейный участок с постоянным наклоном $ d\sigma / d \; ln c_{ПАВ} $ адсорбция достигает постоянного и максимального
значения. В этой области на межфазной границе формируется насыщенный мономолекулярный адсорбционный слой. При дальнейшем увеличении общей концентрации ПАВ ($c_{ПАВ}$ > ККМ) концентрация
молекулярного раствора ПАВ перестает расти, в объеме раствора увеличивается концентрация мицелл,
и поверхностное натяжение остается практически постоянным. Величина ККМ определяется по излому
изотермы при выходе ее на горизонтальный участок.


![рис. 4](image4.png)
##### <center> Рис. 4: Изотерма поверхностного натяжения ПАВ в полулогарифмических координатах <center/>

В работе поверхностное натяжение определяется методом пластины Вильгельми с помощью весов
(рис. 5). В жидкость опускается небольшая пластина, подвешенная к чаше электронных аналитических
весов c жестким подвесом, не опускающимся на пропорциональную нагрузке высоту. Измеряется изменение веса свободно висящей пластины и пластины, погруженной в жидкость. Ёмкость с исследуемым
раствором с помощью специального механизма медленно опускается пока пластинка не выйдет из жидкости - так измеряются показания весов от времени. Также в связи с тем, что в работе используется
шероховатый материал пластинки, покрытие полностью смачивается и $cos \; \theta = 1 $. С учетом этих тонкостей уравнение для разности показаний веса пластинки до и после ее касания поверхности жидкости будет
выглядеть следущим образом:
$$ \Delta F = 2 \sigma \cdot L \qquad (3)$$
где L - ширина пластинки, из этого уравнения следует, что
$$ \sigma = \frac{\Delta m g}{2 L} \qquad (4)$$

#### Определение ККМ в растворе ПАВ кондуктометрическим методом

Для нахождения ККМ измеряли удельную электропроводность раствора ПАВ (в нашем случае SDS)
в зависимости от концентрации. Для этого приготовили 4 раствора с концнтрациями в диапазоне от
0.2ККМ до ККМ и 4 раствора в диапазоне от 1.25ККМ до 3ККМ. Рассчитали объемы жидкости,
необходимые для получения всей серии растворов из первоначального ($C_0$ = 40 мМ) методом замещенного объема. После провели измерения для всех растворов запланированных концентраций, начиная с
минимальной. Повторили измерения, двигаясь в сторону разбавления (замещаемый раствор теперь —
дистиллированная вода). По полученным данным построили график зависимости удельной электропроводности от концентрации ПАВ:

```python
V_working = 22000 
C_0 = 40
KKM = 8.2
```

```python
C1 = np.round((np.concatenate((np.linspace(0.1, 1, 4), np.linspace(1.25, 3, 4)))), 2)
C1
```

```python
res1 = [74.5, 285, 450, 587, 641, 736, 868, 1080]
res2 = [1017, 852, 732, 608, 530, 351, 202, 44.6]
```

```python
fig, ax = plt.subplots(2, figsize=(24, 30))
x = KKM*C1
ax[0].scatter(x, res1)

a1 = np.polyfit(x[:4], res1[:4], deg=1)
f_a1 = np.poly1d(a1)
ax[0].plot(np.linspace(0, 25, 100), f_a1(np.linspace(0, 25, 100)), label=f"Аппроксимирующая прямая: \n y = {a1[0].round(2)}x + {a1[1].round(2)}") 

a2 = np.polyfit(x[4:], res1[4:], deg=1)
f_a2 = np.poly1d(a2)
ax[0].plot(np.linspace(0, 25, 100), f_a2(np.linspace(0, 25, 100)),label=f"Аппроксимирующая прямая: \n y = {a2[0].round(2)}x + {a2[1].round(2)}")
ax[0].axvline((a2[1] - a1[1]) / (a1[0] - a2[0]), c="red", ls="--", label=f"Прямая пересечения x = {np.round((a2[1] - a1[1]) / (a1[0] - a2[0]),2)}")

ax[0].set_xlabel(r"$C_{ПАВ}, М$", fontsize=16)
ax[0].set_ylabel(r"$\kappa$, мкСм/см", fontsize=16)
ax[0].set_title("График зависимости удельной электропроводности от концентрации ПАВ", fontsize=18)

ax[0].grid()
ax[0].legend(loc="best", fontsize=14)


ax[1].scatter(x, res2[::-1])

a1 = np.polyfit(x[:4], res2[::-1][:4], deg=1)
f_a1 = np.poly1d(a1)
ax[1].plot(np.linspace(0, 25, 100), f_a1(np.linspace(0, 25, 100)), label=f"Аппроксимирующая прямая: \n y = {a1[0].round(2)}x + {a1[1].round(2)}") 

a2 = np.polyfit(x[4:], res2[::-1][4:], deg=1)
f_a2 = np.poly1d(a2)
ax[1].plot(np.linspace(0, 25, 100), f_a2(np.linspace(0, 25, 100)),label=f"Аппроксимирующая прямая: \n y = {a2[0].round(2)}x + {a2[1].round(2)}")
ax[1].axvline((a2[1] - a1[1]) / (a1[0] - a2[0]), c="red", ls="--",  label=f"Прямая пересечения x = {np.round((a2[1] - a1[1]) / (a1[0] - a2[0]),2)}")

ax[1].set_xlabel(r"$C_{ПАВ}, М$", fontsize=16)
ax[1].set_ylabel(r"$\kappa$, мкСм/см", fontsize=16)


plt.legend(loc="best", fontsize=14)
plt.grid()
plt.show()
```

При аппроксимации данного графика по точке пересечений двух прямых можем определить приблизительное значение ККМ
$$ ККМ \approx 7.8 \; мМ $$ 
что с не очень хорошей точностью совпадает с табличным значением $ККМ_{theor}$ = 8.2 мМ.

- При концентрации меньше ККМ в растворе присутствуют только одиночные ионы ПАВ. В данном
случае удельная электропроводность выражается как
$$ \kappa = (\lambda_{Br^−} + \lambda_{TTAB})[TTAB] \qquad (5)$$

При концентрации раствора больше ККМ начинается мицеллобразование. Предполагая, что вклад
мицелл в проводимость такой же, что и от эквивалентного количества мономерных катионов, суммарный заряд которых равен заряду мицеллы $ \lambda_{mic} = \alpha N \lambda_{TTAB}$, где $\alpha$ – степень диссоциации мицеллы, $N$ – число агрегации. Тогда получаем следующее выражение для удельной электропроводности:
$$ \kappa = \alpha [TTAB](\lambda_{Br^−} + \lambda_{TTAB}) + (1 − \alpha)(\lambda_{TTAB} + \lambda_{Br^−})[KKM] \qquad (6)$$

Таким образом, по изменению угла наклона прямой можем найти оценить степень ионизации мицелл:
$$ \alpha = \frac{k_2}{k_1} = \frac{28.12}{65.25} \approx 0.431 \qquad (7)$$

#### Определение ККМ методом измерения поверхностного натяжения

Провели измерения поверхностного натяжения в зависимости от концентрации SDS с помощью метода пластины Вильгельми. По полученным данным построили графики зависимости поверхностного
натяжения от концентрации и от логарифма концентрации.

```python
V_working = 30000
l = 1
```

```python
C2 = np.concatenate((np.linspace(0.02, 1, 9), np.linspace(1.25, 2.5, 3)))
C2
```

```python
res3 = np.array([0.1350, 0.1206, 0.1038, 0.0956, 0.0825, 0.0802, 0.0743, 0.0717, 0.0713, 0.0736, 0.0750, 0.0747])
sigma = 100 * 9.8 * res3 / 2*l
sigma
```

```python
fig, ax = plt.subplots(2, figsize=(24, 30))

x = KKM*C2
ax[0].scatter(x, sigma)

ax[0].set_xlabel("C, мM", fontsize=16)
ax[0].set_ylabel(r"$\sigma$, мН/м", fontsize=16)
ax[0].set_title("График зависимости поверхностного натяжения от концентрации", fontsize=18)
ax[0].grid()

ax[1].scatter(np.log(x), sigma)
ax[1].axhline(36.41, ls="--",c="red", label=f"Горизонтальная прямая: \n y = 36.41")
ax[1].axvline(2.10, ls="--", c="g", label=f"Вертикальная прямая: \n x = 2.1")

a1 = np.polyfit(np.log(x[4:9]), sigma[4:9], deg=1)
f_a1 = np.poly1d(a1)
ax[1].plot(np.linspace(-2, 2.5, 100), f_a1(np.linspace(-2, 2.5, 100)), ls="--",label=f"Аппроксимирующая прямая: \n y = {a1[0].round(2)}x + {a1[1].round(2)}") 


ax[1].set_xlabel("ln C", fontsize=16)
ax[1].set_ylabel(r"$\sigma$, мН/м", fontsize=16)
ax[1].set_title("График зависимости поверхностного натяжения от логарифма концентрации", fontsize=18)
ax[1].grid()

plt.legend(loc="best", fontsize=24)
plt.show()
```

По излому кривой из графика определили ККМ, но так как на графике наблюдается минимум,
что может быть связано с наличием поверхностно-активных примесей, точное значение ККМ определить
не удалось. Из нашего графика видим несколько возможных значений ККМ. Мы смогли получить значение
ККМ, близкое к теоретическому: ККМ = 8.17 мМ ($ККМ_{theor}$ = 8.2 мМ), но данный способ не является
точным.

Определили предельную адсорбцию (емкость монослоя, в моль/м2) по наклону изотермы поверхностного натяжения на участке перед ямой на графике:
$$ \Gamma_{\infty} = −\frac{1}{RT} \cdot \frac{d \sigma}{d lnC} = 3.7 \cdot 10^{−6} \; \frac{моль}{м^2} \qquad (8)$$

По найденной величине адсорбции определили площадь, приходящуюся на одну молекулу в плотном
монослое на поверхности раствора:
$$ S_0 = \frac{1}{N_A \Gamma_{\infty}} = 4.5 · 10^{−19} \frac{м^2}{частица} \qquad (9)$$

Задав плотность $d \approx 1 \frac{г}{см^3}$ адсорбированного ПАВ и зная его молярную массу = 288.4 $\frac{г}{моль}$, оценили длину (высоту) молекулы в плотном монослое:
$$ l = \frac{M \Gamma_\infty}{d} = 11 \; Å \qquad (10)$$
Сравнили полученную величину с оценкой по известным геометрическим размерам:
$$ l = 1.256 \cdot (n − 1) + 2 = 1.256 \cdot 11 + 2 = 15.8 \; Å$$

Видим, что экспериментально определённая длина близка к теоретической.
Аналогично определению предельной адсорбции вычислили величины адсорбции в точках изотермы
натяжения до ямы:

Построили изотерму адсорбции и перестроили её в предположении изотермы Ленгмюра в
линеаризующих координатах.

```python
R = 8.31
T = 298.15
```

```python
sigma_ = sigma[1:4]
ln_C = np.log(x[1:4])

d_sigma_d_lnC = np.zeros_like(sigma_)
d_sigma_d_lnC[1:-1] = (sigma_[2:] - sigma_[:-2]) / (ln_C[2:] - ln_C[:-2]) 
Gamma = -d_sigma_d_lnC / (R * T)
```

```python
Gamma
```

```python
df = pd.DataFrame({"C, мМ": x[:4], r"$\Gamma, моль/м^2$": Gamma[:4]}) 
df
```

Табл. 1 Адсорбция от концентрации

```python
fig, ax = plt.subplots(2, figsize=(24, 30))

ax[0].scatter(df["C, мМ"], df["$\Gamma, моль/м^2$"])

ax[0].set_title("График зависимости адсорбции от концентрации", fontsize=20)
ax[0].set_xlabel("C, мМ", fontsize=18)
ax[0].set_ylabel(rf"$\Gamma, мН/м$", fontsize=18)


ax[1].scatter(1 / df["C, мМ"], 1 / df["$\Gamma, моль/м^2$"])

a1 = np.polyfit(1 / df["C, мМ"][1:], 1 / df["$\Gamma, моль/м^2$"][1:], deg=1)
f_a1 = np.poly1d(a1)
ax[1].plot(np.linspace(0, 1, 100), f_a1(np.linspace(0, 1, 100)), ls="--",label=f"Аппроксимирующая прямая: \n y = {a1[0].round(2)}x + {a1[1].round(2)}") 


ax[1].set_title("График зависимости адсорбции от концентрации в линеаризующих координатах", fontsize=20)
ax[1].set_xlabel("1/C, 1/мМ", fontsize=18)
ax[1].set_ylabel(rf"$\Gamma, м^2/моль$", fontsize=18)

plt.legend(loc="best", fontsize=20)
plt.show()
```

По последнему графику определили константу равновесия адсорбции и предельную адсорбцию:
$$ \Gamma = \Gamma_\infty \frac{KC}{1 + KC} \qquad (11)$$
$$ \frac{1}{\Gamma} = \frac{1}{\Gamma_\infty K} \cdot \frac{1}{C} + \frac{1}{\Gamma_\infty} \qquad (12)$$
$$ \Gamma_\infty = 7.9 \cdot 10^{-6} \; \frac{моль}{м^2} \qquad (13)$$
$$ K = 1480 \; \frac{1}{M} \qquad (14)$$

Нашли стандартную свободную энергию адсорбции:
$$ \Delta G^\circ = -RT ln \; K = -18 \; \frac{кДж}{моль} \qquad (15)$$

Сравнили её с оценкой, полученной на основании правила Дюкло-Траубе, в соответствии с которым
поверхностная активность при увеличении углеводородного радикала на одно звено растет в гомологическом ряду ионогенных ПАВ примерно в 2.1 раза:
$$ \frac{g_{n+1}}{g_n} = \frac{\Gamma_{n+1}}{\Gamma_n} = \frac{K_{n+1}}{K_n} = \alpha = 2.1 \qquad (16)$$
$$ \Delta G_{n+1} - \Delta G_n = -RT ln \; \alpha \qquad (17)$$
$$ \Delta G^{\circ}_{298} = -12 \cdot RT ln \; \alpha = -22 \; \frac{кДж}{моль} \qquad(18)$$
Полученный экспериментально результат довольно недалек от теоретического.

### Вывод 

Определение ККМ в растворе ПАВ кондуктометрическим методом показало не очень точные
результаты. При измерениях наибольший вклад в погрешность могло внести приготовление растворов, неточности приборов и расчётов не слишком значительны. Таким образом, кондуктометрический метод не очень подходит для определения ККМ и
он может быть применен только для ионогенных ПАВ.  

Определение ККМ в растворе ПАВ методом определения поверхностного натяжения менее точен, так как очень сложно избавиться от примесей в растворе ПАВ, которые
из-за лучшей адсорбции уменьшают поверхностное натяжение, что приводит к минимуму на
изотерме поверхностного натяжения ПАВ в логарифмических координатах. В результате мы
не имеем чёткого излома, по которому можно точно определить ККМ и можем лишь очень
приблизительно оценивать это значение. Метод определения поверхностного натяжения даёт
очень широкий диапазон значений, в котором может находиться искомое, что делает этот метод
скорее качественным подтверждением теории.   

Так же в процессе работы были произведены некоторые оценки, которые вполне согласуются
с экспериментальными данными, а именно оценка длины молекулы ПАВ в монослое и оценка
стандартной свободной энергии адсорбции. Величина предельной адсорбции, полученная двумя
способами, отличается незначительно.
